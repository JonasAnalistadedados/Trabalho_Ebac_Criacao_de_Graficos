```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('C:/Users/jonat/Downloads/ecommerce_estatistica.csv')
print(df.head().to_string())

# Gr√°fico de Histograma
plt.hist(df['Nota'])
plt.title('Visualizando as Notas')
plt.xlabel('Notas')
plt.ylabel('Quantidade')
plt.show()

# Gr√°fico de Dispers√£o
plt.hexbin(df['Nota'], df['Desconto'], gridsize=40, cmap='Blues')
plt.colorbar(label='Contagem dentro do bin')
plt.xlabel('Nota')
plt.ylabel('Desconto')
plt.title('Dispers√£o de Nota e Desconto')
plt.show()

# Mapa de calor
corr = df[['Nota', 'Pre√ßo']].corr()
plt.subplot(2, 2, 3)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correla√ß√£o Nota e Pre√ßo')
plt.tight_layout()
plt.show()

# Gr√°fico de Barras
plt.figure(figsize=(10, 6))
df['Temporada_Cod'].value_counts().plot(kind='bar', color='#90ee70')
plt.title('Temporada e Quantidade ')
plt.xlabel('Temporada')
plt.ylabel('quantidade')
plt.xticks(rotation=0)
plt.show()

# Gr√°fico de Pizza
x = df['Qtd_Vendidos_Cod'].value_counts().index
y = df['Qtd_Vendidos_Cod'].value_counts().values
plt.figure(figsize=(10,8))
plt.pie(y, labels=x, startangle=70)
plt.title('Distribui√ß√£o de Quantidades Vendidos')
plt.show()

# Gr√°fico Densidade
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Desconto'], fill=True, color='#863e9c')
plt.title('Densidade de Desconto')
plt.xlabel('Desconto')
plt.show()

# Gr√°fico de Regress√£o
sns.regplot(x='Qtd_Vendidos_Cod', y='Temporada_Cod', data=df, color='#278f65', scatter_kws={'alpha': 0.5, 'color': '#34c289'})
plt.title('Regress√£o de Quantidade Vendidos por Temporada')
plt.xlabel('Quantidade Vendidos')
plt.ylabel('Temporada')
plt.show()

```
